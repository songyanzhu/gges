```python
# def func1(x):
#     y = x**2 + x + 1
#     return y
# func1(2)



def func2(x1, x2, constant = 3, constant2 = 1):
    y1 = x1 + 3
    y2 = x1 + x2 + constant + constant2
    if y1 > 5:
        return y1, y2
    else:
        return y1

func2(x2 = 1, x1 = 2, constant2 = 2, constant = 1)
```

```python
def func2(constant = 3, constant2 = 1, x1, x2):
    y1 = x1 + 3
    y2 = x1 + x2 + constant + constant2
    if y1 > 5:
        return y1, y2
    else:
        return y1

# func2(x2 = 1, x1 = 2, constant2 = 2, constant = 1)
func2(1, 2)
```
